#### Visualize the YbCbCr
* RGB Dataset Holopix50k is converted to only YCbCr color space. 
* Only Y channel is fed to SRCNN Model although original paper experiments by training with:
    * RGB channels
    * YCbCr channels
    * Y pre-training
    * CbCr pre-training
    * Bicubic interpol
* Good SR result is obtained on PSNR with both Y only and Y pre-train.

This brief notebook will help visualise the Y Cb and Cr channels

```python
import numpy as np
import PIL.Image as pil 
import os
import torch
import torchvision
import matplotlib.pyplot as plt
import matplotlib.figure as figure

import numpy as np
import skimage.io as io
import torch.utils.data as data
```

```python
train_dir = "91"
input_hr = pil.open(os.path.join(train_dir,'t66.bmp')).convert('RGB')
(input_hr)
```

```python
input_hr = input_hr.convert('YCbCr')
Y = 0
Cb = 1
Cr = 2

YCbCr=list(input_hr.getdata()) # flat list of tuples
# reshape
```

```python
imYCbCr = np.reshape(YCbCr, (input_hr.size[1], input_hr.size[0], 3))
# Convert 32-bit elements to 8-bit
imYCbCr = imYCbCr.astype(np.uint8)

# now, display the 3 channels

temp_negative = pil.fromarray(imYCbCr[:,:,Y], "L").convert('RGB') #convert to RGB to prevent negatives
temp_array = np.asarray(temp_negative)
plt.imshow(temp_array)
```

```python

plt.imshow(pil.fromarray(imYCbCr[:,:,Cb], "L").convert('RGB'))
```

```python
plt.imshow(pil.fromarray(imYCbCr[:,:,Cr], "L").convert('RGB'))
```
